# 1 sorting algorithm

## 1.1 Bubble sort

```python
def BubbleSort(a):
    # wir gehen von einer unsortierten liste aus
    issorted = False
    counter = 0
    # solange nicht sortiert ist, wird folgendes ausgeführt
    
    while not issorted:
        issorted = True              
        # 
        # wenn die range verlassen worden ist, dann wird is sorted auf true gesetzt
        # wir könnten den counter auch vorne auf 1 setzen, und dann nur noch len a minus counter schreiben
        for i in range(len(a) 1 - counter):
            #solange i größer ist als i minus 1 muss geswappt werden.            
            if a[i] > a[i + 1]:
                swap(i, i + 1, a)
                issorted = False
        #wenn die zahl 9 ganz nach hinten gerutscht ist, dann wird der counter hochgesetzt und die range für den nächsten durchgang der if clause gekürzt. 
        counter += 1
    return a

def swap(i, j, a):
    a[i], a[j] = a[j], a[i]

a = [ 9,8,7,6]

BubbleSort(a)

# erst läuft die while loop, automatisch auf false gesetzt
# dann läuft die if range loop, wobei sich diese bei jeden durchgang verkürzt, weil der counter hochgesetzt wird. 
# dann läuft die if clause loop, bis man am ende der reihe angekommen ist. solange die clause läuft issorted false 
```

## 1.2 Selection sort

```python
def Selectionsort(a):
    # current index
    ci = 0
    # solange der current index die range nicht verläßt
    # minus 1, weil wir länge a mit index ci vergleichen
    while ci < len(a) - 1:
        # smallest index
        si = ci
        # die range muss nicht jedes mal von vorne anfangen, da bei jedem durchgang die kleinste zahl nach vorne kommt. 
        for i in range(ci + 1, len(a)):
            # wenn si kleiner ist als i, dann wird der smallest index auf i gesetzt und damit nach rechts verschoben
            if a[si] > a[i]:
                si = i
        # erst wenn die kleinste zahl innerhalb der range ermittelt wurde wird geswappt. 
        swap(ci, si, a)
        #jetzt wird der ci einen hoch gesetzt, weil die kleinste zahl ja schon vorne ist. 
        ci += 1
    return a          
  

def swap(i, j, a):
    a[i], a[j] = a[j], a[i]

a = [ 9,8,7,6]
Selectionsort(a)
```

## 1.3 Insertion Sort

```python
def InsertionSort(a):
    
    # da man j mit j-1 vergleicht, kann die range nicht bei 0 anfangen. 
    for i in range(1, len(a)):
        j = i
        #solange der index j kleiner ist als die zahl davor, muss geswappt werden
        while j > 0 and a[j] < a[j - 1]:
            swap(j, j - 1, a)
            # j minus 1, weil nach jedem swap die zahl davor auch noch geprüft werden muss. bsp. aus 9,8,7 wurde 9,7,8, dann muss jetzt die sieben mit der neun verglichen werden. 
            j -= 1
    return a 
    
    
def swap(i, j, a):
    a[i], a[j] = a[j], a[i]

a = [ 9,8,7,6]
InsertionSort(a)
```
